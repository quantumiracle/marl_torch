## Exploitability test

```python
import numpy as np
fname='20210618_0438.log'  # pong
file = open(fname, 'r')
lines = file.read().splitlines()
print('log indices: ', lines[2])
score_l, length_l = [], []
for l in lines:
    if "Final score" in l:
        length_l.append(int(l.split('episode length: ')[-1]))
        score_l.append(float(l.split(',')[0].split('Final score: ')[-1]))
        print(l)
print(score_l, length_l)
```

```python
import matplotlib.pyplot as plt
plt.plot(score_l)
```

```python
plt.plot(length_l)
```
