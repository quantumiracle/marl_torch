```python
import numpy as np
import matplotlib.pyplot as plt
data = np.load('data/eval_data.npy', allow_pickle=True).item()
print(data)
```

```python
x=[]
r=[]
l=[]
for k,v in data.items():
    x.append(int(k))
    r.append(v[0])
    l.append(v[1])
```

```python
plt.xlabel('training episode')
plt.ylabel('average eval reward')
plt.grid()
plt.plot(x, r)
```

```python
plt.xlabel('training episode')
plt.ylabel('average episode length')
plt.grid()
plt.plot(x, l)
```
